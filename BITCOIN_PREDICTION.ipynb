BITCOIN PREDICTION

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

LOAD DATASET


```python
df=pd.read_csv('BitcoinPrice.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.drop(['Date'],axis=1,inplace=True)
```

```python
df.head()
```

PREDICTING 30DAYS IN FUTURE

```python
p_days=30
```

```python
df['Prediction']=df[['Price']].shift(-p_days)
```

```python
df.head()
```

```python
df.tail()
```

SEPARATE X AND Y

```python
X=np.array(df.drop(['Prediction'],1))
```

```python
X=X[: len(df)-p_days]
```

```python
X.shape
```

```python
X
```

CREATING Y AS NP ARRAY

```python
Y = np.array(df['Prediction'])
Y=Y[: -p_days]
```

```python
Y.shape
```

```python
Y
```

SPLIT THE DATA

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size =0.2)
```

```python
P_days_array = np.array(df.drop(['Prediction'],1))[-p_days:]
```

```python
P_days_array
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
Rf = RandomForestRegressor(n_estimators=1000,random_state=1)
```

```python
Rf.fit(X_train,y_train)
```

```python
print('Random Forest Accuracy: {:.2f}%'.format(Rf.score(X_test,y_test)*100))
```

PREDICTION

```python
Rf_predict =Rf.predict(X_test)
```

```python
Rf_predict
```

```python
y_test
```

MODEL PREDICTION FOR DAYS

```python
Rf_predict_30= Rf.predict(P_days_array)
Rf_predict
```
